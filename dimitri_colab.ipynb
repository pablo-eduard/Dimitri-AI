## 2️⃣ Configurar Variáveis de Ambiente

```python
import os
from getpass import getpass

# Solicita as credenciais de forma segura
print("🔐 Configure suas credenciais (não serão salvas em histórico):")
print()

bluesky_handle = input("Seu handle do Bluesky (ex: dimitriai.bsky.social): ").strip()
bluesky_password = getpass("Senha de app do Bluesky: ")
admin_handle = input("Handle do admin (seu próprio handle): ").strip()
hf_api_key = getpass("Chave de API do Hugging Face (hf_xxx...): ")

# Define variáveis de ambiente
os.environ["BLUESKY_HANDLE"] = bluesky_handle
os.environ["BLUESKY_APP_PASSWORD"] = bluesky_password
os.environ["ADMIN_HANDLE"] = admin_handle
os.environ["HUGGINGFACE_API_KEY"] = hf_api_key

print("\n✅ Variáveis configuradas!")
```

## 4️⃣ Testar Conexão com Bluesky

```python
from atproto import Client

try:
    client = Client()
    profile = client.login(
        os.environ["BLUESKY_HANDLE"],
        os.environ["BLUESKY_APP_PASSWORD"]
    )
    print(f"✅ Conectado ao Bluesky como {profile.handle}")
    print(f"   Display name: {profile.display_name}")
except Exception as e:
    print(f"❌ Erro ao conectar: {e}")
```

## 5️⃣ Testar API do Hugging Face (Qwen)

```python
import requests

HF_API_KEY = os.environ.get("HUGGINGFACE_API_KEY")
MODELO_QWEN = "Qwen/Qwen0.5B"
HF_API_URL = f"https://api-inference.huggingface.co/models/{MODELO_QWEN}"

headers = {
    "Authorization": f"Bearer {HF_API_KEY}",
    "Content-Type": "application/json",
}

payload = {
    "inputs": "Olá! Como você está?",
    "parameters": {
        "max_new_tokens": 50,
        "temperature": 0.7,
    },
}

try:
    response = requests.post(HF_API_URL, headers=headers, json=payload, timeout=30)
    if response.status_code == 200:
        resultado = response.json()
        print(f"✅ Qwen respondeu:")
        print(f"   {resultado[0]['generated_text']}")
    else:
        print(f"⚠️ Status {response.status_code}: {response.text}")
except Exception as e:
    print(f"❌ Erro: {e}")
```
